## This notebook is for preprocessing files, including extraction of images (a.k.a. frames) from a video file and resizing them to standard format

```python
## imports
import cv2 
from PIL import Image
import os, sys
import datetime 
```

### Here are configurations

```python
## configuration inputs
width=200
height=150

## path - the folder where your video files reside 
input_path = "C:\\backup\\241\\videos"
output_path = input_path + "Result\\"
output_path_resized = input_path + "Result_Resized\\"
pattern=".mp4"
filename="Sec5_10292020_batch2_700_aftertreat.mp4"
```

### Extraction of frames

```python

#dirs = os.listdir( input_path )
#cwd = os.getcwd()
cwd = os.listdir( input_path )

#create Result dir if not exist
if not os.path.isdir(output_path):
    os.mkdir(output_path)

#traverse through every mp4 file in cwd
for file in cwd:
    if file.endswith(pattern):
        fileName = os.path.join(input_path, file) #mp4 file working on
        vidcap = cv2.VideoCapture(fileName)
        success,image = vidcap.read()
        count = 0
        success = True
        while success:
          vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*1000)) 
          success,image = vidcap.read()
          print ('Read a new frame: ', success)
          outputPath = input_path+ "Result\\{}_frame{}.jpg"
          name,ff = file.split(".")
          cv2.imwrite( outputPath.format(name,count), image) # save frame as JPEG file
          print(outputPath.format(name,count))
          count = count + 1
```

### Resizing of images

```python
#create Result dir if not exist
if not os.path.isdir(output_path_resized):
    os.mkdir(output_path_resized)

#path = "C:\\Users\\alexa\\Downloads\\Result\\"
dirs = os.listdir( output_path) 

print ('Directory: ', output_path_resized)

for item in dirs:
        print ('Directory: ', output_path_resized+item)
        if os.path.isfile(output_path+item):
            print ('Resizing: ', item)       
            im = Image.open(output_path+item)
            f, e = os.path.splitext(output_path_resized+item)
            imResize = im.resize((width,height), Image.ANTIALIAS)
            imResize.save(f + ' resized.jpg', 'JPEG', quality=90)

 
```

## Time-based naming of frames

```python
filename="Sec5_10292020_batch1_640_pretreat.mp4"
```

```python
def GetTimeFromFileName(filename):
    piece = filename.split("_")
    date = dt.datetime.strptime(piece[1], '%m%d%Y')
    time=dt.datetime.strptime(piece[3], '%H%M')
    correction = -5 if piece[4].startswith("pretreat") else 7
    hours_added = dt.timedelta(hours = time.hour+12)
    minutes_added = dt.timedelta(minutes = time.minute+correction)
    seconds_added = dt.timedelta(seconds = 0)       
    actual_date_and_time = date + hours_added + minutes_added +seconds_added
    return (piece[0],actual_date_and_time)
    
```

```python
def GetFramesFromVideoFile(input_path, file):
    sec,timing=GetTimeFromFileName(file)
    #fileName = os.path.join(input_path, file) #mp4 file working on
    fileName = input_path+"\\"+file
    #print(input_path)
    print(fileName)   
    vidcap = cv2.VideoCapture(fileName)
    success,image = vidcap.read()
    count = 0
    #success = True
    target_folder,ff = file.split(".")   
    target_path=input_path+"\\"+target_folder
    if not os.path.exists(target_path):
        os.makedirs(target_path)
    while success:
      vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*1000)) 
      print ('Read a new frame: ', success, type(image), image.shape)
      #outputPath = input_path+ "\\Result\\{}_frame{}.PNG"         
      outputPath = target_path+ "\\{}_frame_{}.JPG"
      seconds_added = dt.timedelta(seconds = count)                                            
      date_and_time = timing + seconds_added
      #name=sec+date_and_time.isoformat(' ')
      # out_file=outputPath.format(sec,date_and_time.isoformat(' '))
      out_file=outputPath.format(sec,date_and_time.strftime("%d_%m_%Y_%H_%M_%S"))
      cv2.imwrite( out_file, image)  
      print(out_file)
      count = count + 1
      success,image = vidcap.read()
      
```
